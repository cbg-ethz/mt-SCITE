```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
scores = pd.read_csv('../../data/YFV2001_learn_error_rate/yfv2001_scores_cv.csv')
scores = scores.drop('Unnamed: 0', axis=1)
scores = scores.sort_index(axis=1)
```

```python
scores
```

```python
# Create a boxplot
plt.figure(figsize=(12, 6)) 
sns.boxplot(data=scores, orient='vertical', palette='Set1')
plt.xticks(rotation=90)  
plt.xlabel('Error rate')
plt.ylabel('normalized lhood')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
pd.set_option('display.max_columns', None)

variance_row = scores.iloc[0:3, :].var(axis=0)
mean_row = scores.iloc[0:3, :].mean(axis=0)
median_row = scores.iloc[0:3, :].median(axis=0)

# Calculate median/variance
median_over_variance_row = median_row / variance_row
mean_over_variance_row = mean_row / variance_row

# Create new rows and add them to the DataFrame
scores.loc['variance'] = variance_row
scores.loc['mean'] = mean_row
scores.loc['median'] = median_row
scores.loc['median/variance'] = median_over_variance_row
scores.loc['mean/variance'] = mean_over_variance_row
scores
```

```python
# Calculate mean and variance for each column
mean_values = scores.mean()
variance_values = scores.var()

# Calculate standard deviation (sigma) from variance
std_deviation_values = np.sqrt(variance_values)

# Calculate the tradeoff score (Mean / Standard Deviation) for each column
tradeoff_scores = mean_values / std_deviation_values

# Rank the columns based on tradeoff scores (higher is better)
ranked_columns = tradeoff_scores.sort_values(ascending=False)

# Select the column with the highest tradeoff score as the "best" column
best_column = ranked_columns.index[0]

# Print the best column
print("The best column based on the tradeoff between mean and variance:", best_column)
```

```python
# Calculate mean and variance for each column
mean_values = scores.mean()
variance_values = scores.var()

# Create a DataFrame to hold mean and variance information
mean_var_df = pd.DataFrame({'Mean': mean_values, 'Variance': variance_values})
#print(mean_var_df)

# Sort the DataFrame by Mean in descending order
mean_var_df_sorted = mean_var_df.sort_values(by='Mean', ascending=False)

# Find the highest mean value
highest_mean = mean_var_df_sorted['Mean'].max()

# Filter the DataFrame to select columns with the highest mean
highest_mean_columns = mean_var_df_sorted[mean_var_df_sorted['Mean'] == highest_mean]
#print(highest_mean_columns)

# Find the column with the lowest variance among the highest mean columns
best_column = highest_mean_columns[highest_mean_columns['Variance'] == highest_mean_columns['Variance'].min()]

# Print the best column
if not best_column.empty:
    print("Column with the highest mean and lowest variance:")
    print(best_column)
else:
    print("No column found with the highest mean and lowest variance.")
```

```python
# Calculate mean and variance for each column
mean_values = scores.mean()
variance_values = scores.var()

# Create a DataFrame to hold mean and variance information
mean_var_df = pd.DataFrame({'Mean': mean_values, 'Variance': variance_values})

# Calculate the threshold for the top 10% of means
top_10_percent_threshold = mean_var_df['Mean'].quantile(0.9)

# Filter the DataFrame to select columns with means in the top 10%
top_10_percent_columns = mean_var_df[mean_var_df['Mean'] >= top_10_percent_threshold]

# Find the column with the lowest variance among the top 10% mean columns
best_column = top_10_percent_columns[top_10_percent_columns['Variance'] == top_10_percent_columns['Variance'].min()]

# Print the best column
if not best_column.empty:
    print("Column with the highest mean and lowest variance among the top 10% mean columns:")
    print(best_column)
else:
    print("No column found with the highest mean and lowest variance among the top 10% mean columns.")
```
